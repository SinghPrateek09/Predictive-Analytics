```python
# Importing necessary libraries 

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from  sklearn.preprocessing import StandardScaler
from sklearn import tree
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import statsmodels.api as sm
from sklearn.ensemble import RandomForestClassifier
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

# ::--------------------------- Exploratory Data Analysis ---------------------------- ::

```python
# Loading Personal Bank Loan dataset

Subs = pd.read_excel('marketing_campaign.xlsx')
Subs
```

```python
# Exploratory Analysis and Summary of the Dataset

Subs.info()
```

```python
# Number of records summary of the Dataset

NoOfRows, NoOfCols = Subs.shape
print('Total Number of rows :', NoOfRows)
print('Total Number of columns :', NoOfCols)
```

```python
# Checking for the missing or null values

Subs.isnull().sum()
```

```python
# Finding the unique value of Education level where the Income is null

print('Missing values by Education = ', Subs[Subs['Income'].isna()]['Education'].unique())
```

```python
# Getting the median values of Income grouped by Education Level 

Subs.groupby('Education').median()['Income'].reset_index()
```

```python
# Replacing the values of Null Income with corresponding Education Level median

def replace_null(col):
    if pd.isna(col['Income']):
        if col['Education'] == '2n Cycle':
            return 46805.0
        elif col['Education'] == 'Graduation':
            return 52028.5
        elif col['Education'] == 'PhD':
            return 55212.0
        elif col['Education'] == 'Master':
            return 50943.0
    else:
        return col['Income']
```

```python
Subs['Income'] = Subs[['Education','Income']].apply(replace_null, axis = 1)
```

```python
# Checking for the missing or null values

Subs.isnull().sum() 
```

```python
# Drop unneccesary columns

Subs.drop(['ID','Dt_Customer', 'Z_CostContact', 'Z_Revenue'], axis = 1, inplace = True)
```

```python
# Grouping Amount Spent on several products into single column

Subs["Spent"] = Subs["MntWines"]+ Subs["MntFruits"]+ Subs["MntMeatProducts"]+ Subs["MntFishProducts"]+ Subs["MntSweetProducts"]+ Subs["MntGoldProds"]
```

```python
# Dropping Replaced Columns

Subs.drop(['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'], axis = 1, inplace = True)
```

```python
# Grouping Marital status into two categories of "Having a Partner" and "Alone"

Subs["Marital_Status"] = Subs["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone"})
```

```python
# Grouping Children as Total Children

Subs["Children"] = Subs["Kidhome"] + Subs["Teenhome"]
```

```python
# Dropping Replaced Columns

Subs.drop(['Kidhome', 'Teenhome'], axis = 1, inplace = True)
```

```python
# Grouping Education into lesser categories

Subs["Education"] = Subs["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Graduate", "PhD":"PhD"})
```

```python
# Grouping Purchases into Total purchases

Subs["Purchase"] = Subs["NumDealsPurchases"]+ Subs["NumWebPurchases"]+ Subs["NumCatalogPurchases"]+ Subs["NumStorePurchases"]
```

```python
# Dropping Replaced Columns

Subs.drop(['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases'], axis = 1, inplace = True)
```

```python
# Displaying cleaned dataset info

Subs.info()
```

```python
# Converting Categorical Variables to numerical values

LE = LabelEncoder()
Subs['Education'] = Subs[['Education']].apply(LE.fit_transform)
Subs['Marital_Status'] = Subs[['Marital_Status']].apply(LE.fit_transform)
```

# ::------------------------- Building Logistic Regression Model ----------------------------::

```python
# Checking multicollinearity

Const = sm.tools.add_constant(Subs) 
VIF_Results = pd.Series([variance_inflation_factor(Const.values,i) for i in range(Const.shape[1])],
                  index = Const.columns)
VIF_Results
```

```python
# Splitting Dataset into Train and Test Set for Target(Dependent) and Independent Variables

np.random.seed(42)

Target_Var = Subs[['Response']]
Independent_Var = Subs.iloc[:,:-1]
Independent_Var_train, Independent_Var_test, Target_Var_train, Target_Var_test = train_test_split(Independent_Var, Target_Var, test_size = 0.3, random_state = 30)
```

```python
# Creating Logistic Regression Model

LR = LogisticRegression()
LR.fit(Independent_Var_train, Target_Var_train)
```

```python
# To get a Logistic Regression Model summary

Model = sm.Logit(Target_Var_train, Independent_Var_train)
X = Model.fit()
print(X.summary())
```

```python
# Prediction over Test Samples

LR_pred = LR.predict(Independent_Var_test)
```

```python
# Confusion matrix

LR_Confusion_Matrix = confusion_matrix(Target_Var_test, LR_pred)

print('\nLogistic Regression Confusion Matrix: \n', LR_Confusion_Matrix)
print('\nTrue Positive    = ', LR_Confusion_Matrix[1][1])
print('True Negative     = ',   LR_Confusion_Matrix[0][0])
print('False Positive   = ',   LR_Confusion_Matrix[0][1])
print('False Negative    = ',   LR_Confusion_Matrix[1][0])
 
# accuracy score of the model
print('\nTest accuracy = ', accuracy_score(Target_Var_test, LR_pred))

# precision of the model
print('\nTest Precision = ', precision_score(Target_Var_test, LR_pred))
```

```python
# Plotting the Confusion Matrix

plt.figure(figsize=(7,7))
sns.heatmap(LR_Confusion_Matrix, annot = True, linewidths = .5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(accuracy_score(Target_Var_test, LR_pred))
plt.title(all_sample_title, size = 15);
```

# ::---------------------- Building Support Vector Machine Model -----------------------::

```python
# Creating SVM Model

SVM = SVC()
SVM.fit(Independent_Var_train, Target_Var_train)
```

```python
# Prediction over Test Samples

SVM_pred = SVM.predict(Independent_Var_test)
```

```python
# Plotting the Confusion Matrix

plt.figure(figsize=(7,7))
sns.heatmap(SVM_Confusion_Matrix, annot = True, linewidths = .5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(accuracy_score(Target_Var_test, SVM_pred))
plt.title(all_sample_title, size = 15);
```
