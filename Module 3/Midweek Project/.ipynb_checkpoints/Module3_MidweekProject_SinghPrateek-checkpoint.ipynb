```python
# Loading Personal Bank Loan dataset

Loans = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx', sheet_name='Data')
Loans
```

```python
sns.pairplot(Loans.iloc[:,1:])
```

```python
# Replacing Negative Experience values with median values of positive experience values of the particular age group and education level

Loans_Pos_Exp = Loans[Loans['Experience'] > 0]
Loans_Neg_Exp = Loans[Loans['Experience'] < 0]
Loans_Neg_Exp_List = Loans_Neg_Exp['ID'].tolist()

for id in Loans_Neg_Exp_List:
    age = Loans.loc[np.where(Loans['ID'] == id)]["Age"].tolist()[0]
    edu = Loans.loc[np.where(Loans['ID'] == id)]["Education"].tolist()[0]
    pos_exp_filter = Loans_Pos_Exp[(Loans_Pos_Exp['Age'] == age) & (Loans_Pos_Exp['Education'] == edu)]
    if pos_exp_filter.empty :
        neg_exp_filter = Loans_Neg_Exp[(Loans_Neg_Exp['Age'] == age) & (Loans_Neg_Exp['Education'] == edu)]
        Exp = round(neg_exp_filter['Experience'].median())
    else:
        Exp = round(pos_exp_filter['Experience'].median())
    Loans.loc[Loans.loc[np.where(Loans['ID'] == id)].index, 'Experience'] = abs(Exp)
```

# ::------------------------- Building Logistic Regression Model ---------------------------------::

```python
Loans = Loans.drop(['ID','ZIP Code'], axis=1)
```

```python
# Checking multicolilinearity

Const = sm.tools.add_constant(Loans) 
VIF_Results = pd.Series([variance_inflation_factor(Const.values,i) for i in range(Const.shape[1])],
                  index = Const.columns)
VIF_Results
```

```python
# Splitting Dataset into Train and Test Set for Target(Dependent) and Independent Variables

np.random.seed(42)

Target_Var = Loans[['Personal Loan']]
Independent_Var = Loans.iloc[:,:-1]
Independent_Var_train, Independent_Var_test, Target_Var_train, Target_Var_test = train_test_split(Independent_Var, Target_Var, test_size = 0.3, random_state = 30)
```

```python
# Creating Logistic Regression Model

Model = sm.Logit(Target_Var_train, Independent_Var_train)
Result = Model.fit()
print(Result.summary())
```

```python
print(Result.summary2())
```

```python
# performing predictions on the test datdaset
yhat = Result.predict(Independent_Var_test)
LgR_Model_Predict = list(map(round, yhat))
 
# comparing original and predicted values of y
print('Actual values', list(Target_Var_test.values))
print('Predictions :', LgR_Model_Predict)
```

```python
# confusion matrix
LgR_Confusion_Matrix = confusion_matrix(Target_Var_test, LgR_Model_Predict)

print('\nLogistic Regression Confusion Matrix: \n', LgR_Confusion_Matrix)
print('\nTrue Positive    = ', LgR_Confusion_Matrix[1][1])
print('True Negative     = ',   LgR_Confusion_Matrix[0][0])
print('False Positive   = ',   LgR_Confusion_Matrix[0][1])
print('False Negative    = ',   LgR_Confusion_Matrix[1][0])
 
# accuracy score of the model
print('\nTest accuracy = ', accuracy_score(Target_Var_test, LgR_Model_Predict))


```
