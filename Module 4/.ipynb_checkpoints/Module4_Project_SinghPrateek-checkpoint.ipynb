# ::------------------------- Building Linear Regression Model ----------------------------::

```python
# Splitting Dataset into Train and Test Set for Target(Dependent) and Independent Variables

np.random.seed(42)

Target_Var = Nash_Housing[['Sale Price']]
Independent_Var = Nash_Housing.iloc[:,:-1]
Independent_Var_train, Independent_Var_test, Target_Var_train, Target_Var_test = train_test_split(Independent_Var, Target_Var, test_size = 0.3, random_state = 30)
```

```python
# Building a Linear Regression Model using all the variables

Constant_Var = sm.add_constant(Independent_Var_train)
Est = sm.OLS(Target_Var_train, Constant_Var)
Est2 = Est.fit()
print(Est2.summary())
```

```python
# Getting Best Variables for a Linear Regression Model using Forward Selection Method

def forward_regression(Independent_Var_train, Target_Var_train,
                       threshold_in = 0.05,
                       verbose = False):
    feature_list = []
    included = list(feature_list)
    while True:
        changed = False
        excluded = list(set(Independent_Var_train.columns)-set(included))
        new_pval = pd.Series(index = excluded)
        for new_column in excluded:
            model = sm.OLS(Target_Var_train, sm.add_constant(pd.DataFrame(Independent_Var_train[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < threshold_in:
           best_feature = new_pval.idxmin()
           included.append(best_feature)
           changed = True
           if verbose:
                print('Add with p-value '.format(best_feature, best_pval))

        if not changed:
            break

    return included

forward_regression(Independent_Var_train, Target_Var_train)
```

```python
# Dropping insignificant variables from Independent Train to re-test Linear Regression Model

Independent_Var_train.drop('Acreage', axis = 1, inplace = True)
Independent_Var_train.drop('Building Value', axis = 1, inplace = True)
Independent_Var_train.drop('Finished Area', axis = 1, inplace = True)
Independent_Var_train.drop('Foundation Type', axis = 1, inplace = True)
Independent_Var_train.drop('Exterior Wall', axis = 1, inplace = True)
Independent_Var_train.drop('Grade', axis = 1, inplace = True)
Independent_Var_train.drop('Bedrooms', axis = 1, inplace = True)
Independent_Var_train.drop('Full Bath', axis = 1, inplace = True)
Independent_Var_train.drop('Half Bath', axis = 1, inplace = True)
Independent_Var_train.drop('Sale Month', axis = 1, inplace = True)
Independent_Var_train.drop('Bldg Age', axis = 1, inplace = True)
Independent_Var_train
```

```python
# Dropping insignificant variables from Independent Test to re-test Linear Regression Model

Independent_Var_test.drop('Acreage', axis = 1, inplace = True)
Independent_Var_test.drop('Building Value', axis = 1, inplace = True)
Independent_Var_test.drop('Finished Area', axis = 1, inplace = True)
Independent_Var_test.drop('Foundation Type', axis = 1, inplace = True)
Independent_Var_test.drop('Exterior Wall', axis = 1, inplace = True)
Independent_Var_test.drop('Grade', axis = 1, inplace = True)
Independent_Var_test.drop('Bedrooms', axis = 1, inplace = True)
Independent_Var_test.drop('Full Bath', axis = 1, inplace = True)
Independent_Var_test.drop('Half Bath', axis = 1, inplace = True)
Independent_Var_test.drop('Sale Month', axis = 1, inplace = True)
Independent_Var_test.drop('Bldg Age', axis = 1, inplace = True)
Independent_Var_test
```

```python
# Building a Linear Regression Model using the best significant variables

Constant_Var = sm.add_constant(Independent_Var_train)
Est = sm.OLS(Target_Var_train, Constant_Var)
Est2 = Est.fit()
print(Est2.summary())
```

```python
# Fitting a Linear Regression Model on training data set

linear_model = LinearRegression().fit(Independent_Var_train, Target_Var_train)

# make predictions

LM_Pred = linear_model.predict(Independent_Var_test)
```

```python
# Displaying the predicted values for test model

print('Coefficients:', linear_model.coef_)
print('Intercept:', linear_model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Target_Var_test, LM_Pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Target_Var_test, LM_Pred))
print('Root Mean Squared Error: %.2f'
      % np.sqrt(metrics.mean_squared_error(Target_Var_test, LM_Pred)))
```
