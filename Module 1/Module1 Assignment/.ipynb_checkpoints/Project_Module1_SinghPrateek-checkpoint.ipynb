```python
# Transform the data set into Independant variable dataset and Target variable dataset

np.random.seed(42)

# Initializing the sample size (~10% of actual dataset) and creating sample dataset
Sample_Size = 5000
Sample_Income_Data = Income_Data.sample(n = Sample_Size)


# Sepearting Independant variable dataset and Target variable dataset

Sample_Income_Data_x, Sample_Income_Data_y = Sample_Income_Data.iloc[:,:-1], Sample_Income_Data.iloc[:,-1]

Sample_Income_Data_x = Sample_Income_Data_x.to_numpy()
print(Sample_Income_Data_x)

Sample_Income_Data_y = Sample_Income_Data_y.to_numpy()
print(Sample_Income_Data_y)
```

```python
# Creating a Training Set Data, Test Set Data, Training Set Labels and Test Set Labels. Test Set size is kept at 30% of the entire Sample.

Training_data, Test_data, Training_labels, Test_labels = train_test_split(Sample_Income_Data_x, Sample_Income_Data_y, test_size = 0.30, random_state = 25)


print(Training_data)
print(Training_labels)
print(Test_data)
print(Test_labels)
print(len(Training_data))
print(len(Test_data))
```

```python
# Plotting the two Labels of Income earned (above 50K or below 50K) in Census Dataset and see their distribution.

X = []
for iclass in range(2):
    X.append([[],[]])
    for i in range(len(Training_data)):
        if Training_labels[i] == iclass:
            X[iclass][0].append(Training_data[i][0])
            X[iclass][1].append(sum(Training_data[i][1:]))
            
print(X)

colours = ("r","g")

plot = plt.figure()
ax = plot.add_subplot(111, projection = '3d')

for iclass in range(2):
    ax.scatter(X[iclass][0], X[iclass][1], c = colours[iclass])

plt.show()
```

```python
# Calculating the KNN or the K nearest neighbour distance

def distance(point1, point2):
    
    point1 = np.array(point1)
    point2 = np.array(point2)
    
    return np.linalg.norm(point1 - point2)
```

```python
# Building KNN Model to correctly calculate the distance and predict the random datapoint how it can be classified as

def get_neighbor(training_set, labels, test_instance, k, distance = distance):
    
    distances = []
    for index in range(len(training_set)):
        dist = distance(test_instance, training_set[index])
        distances.append((training_set[index], dist, labels[index]))
    distances.sort(key = lambda x: x[1])
    neighbors = distances[:k]
    return(neighbors)
```

```python
# Determine which are the closest datapoints to the test value which is getting predicted

def vote(neighbors):
    class_counter = Counter()
    for neighbor in neighbors:
        class_counter[neighbor[2]] += 1
    return class_counter.most_common(1)[0][0]
```

```python
# Display vote and the labels for the Test set data points and see the accuracy of the model

var = 0
for i in range(len(Test_data)):
    neighbors = get_neighbor(Training_data, Training_labels, Test_data[i], 38, distance = distance)
    print("index:", i, ",result of vote:", vote(neighbors), ", label:", Test_labels[i], ",data:", Test_data[i])
    if(vote(neighbors) != Test_labels[i]):
        var += 1
    
print("Accuracy of the model is " + str((1-(var/len(Test_data)))*100) + "% for k = 38")
```
